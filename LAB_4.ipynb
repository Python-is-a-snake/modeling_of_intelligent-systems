# Варіант 3

\[
\begin{cases}
8.2x_1 + 7.9x_2 &= 123 \\
11.2x_1 + 16.3x_2 &= 201 \\
-16.3x_1 - 8.3x_2 &= 169 \\
\end{cases}
\]


```python
import numpy as np
import matplotlib.pyplot as plt

x2 = np.linspace(0, 20, 100)

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 5), sharey=True, dpi=1080)

f1 = (123 - 7.9 * x2) / 8.2
f2 = (201 - 16.3 * x2) / 11.2
f3 = (169 + 8.3 * x2) / (-16.3)

ax[0].plot(x2, f1, 'b-.')
ax[0].grid(True, linestyle='--', color='gray')
ax[0].set(xlim=(0, 20), ylim=(-20, 30), title='$8.2x_1 + 7.9x_2 = 123$')

ax[1].plot(x2, f2, 'r:')
ax[1].grid(True, linestyle='--', color='gray')
ax[1].set(xlim=(0, 20), title='$11.2x_1 + 16.3x_2 = 201$')

ax[2].plot(x2, f3, color='Green', linestyle='-')
ax[2].grid(True, linestyle='--', color='gray')
ax[2].set(xlim=(0, 20), title='$-16.3x_1 - 8.3x_2 = 169$')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x_values = np.linspace(-10, 100, 100)

# Calculate functions
func_1 = (123 - 7.9 * x_values) / 8.2
func_2 = (201 - 16.3 * x_values) / 11.2
func_3 = (169 + 8.3 * x_values) / (-16.3)

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(12, 8), dpi=1080)

# Set title and grid for the plot
ax.set_title('All lines in one figure')
ax.grid(True, linestyle='--', color='gray')

# Plot lines and add text labels
ax.plot(x_values, func_1, 'b-.', label='$8.2x_1 + 7.9x_2 = 123$')
ax.text(x_values[26], func_1[25], '$8.2x_1 + 7.9x_2 = 123$', rotation=-np.arctan(7.9/8.2)*180/np.pi + 7, ha='right', va='bottom', color='black')

ax.plot(x_values, func_2, 'r:', label='$11.2x_1 + 16.3x_2 = 201$')
ax.text(x_values[47], func_2[46], '$11.2x_1 + 16.3x_2 = 201$', rotation=-np.arctan(16.3/11.2)*180/np.pi + 7, ha='right', va='bottom', color='black')

ax.plot(x_values, func_3, color='Green', linestyle='-', label='$-16.3x_1 - 8.3x_2 = 169$')
ax.text(x_values[25], func_3[24], '$-16.3x_1 - 8.3x_2 = 169$', rotation=-np.arctan(-8.3/-16.3)*180/np.pi + 7, ha='right', va='bottom', color='black')

# Set limits for x and y axes
ax.set_xlim(0, 70)
ax.set_ylim(-45, 15)

# Find intersection points
coeff_12 = np.array([[8.2, 7.9], [11.2, 16.3]])
const_12 = np.array([123, 201])
inter_12 = np.linalg.solve(coeff_12, const_12)

coeff_13 = np.array([[8.2, 7.9], [-16.3, -8.3]])
const_13 = np.array([123, 169])
inter_13 = np.linalg.solve(coeff_13, const_13)

coeff_23 = np.array([[11.2, 16.3], [-16.3, -8.3]])
const_23 = np.array([201, 169])
inter_23 = np.linalg.solve(coeff_23, const_23)

intersection_points = np.array([inter_12, inter_13, inter_23])
colors = np.array(['r', 'g', 'b'])

# Plot intersection points and add annotations
for i, point in enumerate(intersection_points):
    ax.scatter(point[1], point[0], marker='o', color=colors[i], alpha=1, zorder=10)
    ax.annotate(f'Intersection {i + 1}', xy=(point[1], point[0]), xytext=(point[1] + 10, point[0] + 10),
                arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=8, ha='left', va='bottom')

# Sort points by x values
sorted_points = intersection_points[intersection_points[:, 1].argsort()]

# Fill the triangle area
plt.fill_between(x_values, func_1, func_2, where=(x_values >= sorted_points[0][1]) & (x_values <= sorted_points[1][1] + 1), color='gray', alpha=0.3)
plt.fill_between(x_values, func_1, func_3, where=(x_values >= sorted_points[1][1]) & (x_values <= sorted_points[2][1]), color='gray', alpha=0.3)

# Add legend
lg = ax.legend()

# Save the plot in different formats
plt.savefig('figure.jpg', format='jpg')
plt.savefig('figure.svg', format='svg')
plt.savefig('figure.png', format='png')

# Display the plot
plt.show()
```
